```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
#G = nx.DiGraph()
```

```python
City_set = ['Mumbai', 'Hyderabad' , 'kolkata' , 'Chennai' , 'Delhi']
```

```python
for each in City_set:
    G.add_node(each)
```

```python
nx.draw(G)
```

```python
nx.draw(G,with_labels=1)
```

```python
G.add_edge('Delhi' , 'kolkata')
```

```python
G.add_edge('Hyderabad' , 'Delhi')
```

```python
G.add_edge('Hyderabad' , 'Chennai')
```

```python
G.add_edge('Chennai', 'Mumbai')
```

```python
G.add_edge('Mumbai', 'kolkata')
```
